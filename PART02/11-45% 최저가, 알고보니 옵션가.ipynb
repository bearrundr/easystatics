## 45% 최저가, 알고보니 옵션가 (12일차) -- 조건부

   * 데이터 속에 있는 차이를 설명하거나 찾을려면, 변수들의 관계를 잘 활용하여야 합니다.
   * 특히 조건이 붙은 변수들은 평균도 확률도 확 바뀝니다.
   * 예를 들면 평균년봉을 말할때도 대기업 평균연봉과 전체 평균연봉은 다릅니다. 대기업이란 조건이 붙기 때문입니다.
   * 데이터의 이런 성질을 이용하여 데이터의 차이를 찾아내고 설명 할 수 있습니다.
   * 이떄 년봉을 **관심변수(종속변수**), 대기업 같은 조직을 **설명변수(독립변수)**라고 합니다.

## 조건부 확률과 조건부 평균

   * 아빠-아들 키 데이터에서 아들키 데이터를 분석하려고 합니다.
       * 누군가는 $150 cm$ 도 안되고 누군가는 $200 cm$ 가 넘습니다.
       * 이런 키 차이를 설명할 변수의 특징을 찾아야 합니다.
       * 그러나, 아들 변수 자체로는 변수의 특징을 알 수는 있어도, 키 차이가 나오는지는 설명하지 못합니다.
       * 즉 관심변수(아들키) 는 평균이나, 다섯숫자 요약을 통해서 변수의 특징을 설명하는 것이 최선입니다.
       * 키차이가 나는 이유를 설명하는 다른 변수가 필요합니다.

   * 앞서 구한 아들키와 아빠키의 상관계수는 0.5 입니다.
   * 0.5 라는 숫자는 아빠키가 크면 아들키도 크는데 영향이 있다고 합니다.

#### 조건부 확률, 조건부 평균

   * 아빠 키 라는 설명변수를 가지고 아들키를 분석해 봅니다.
   * 아들키가 $180 cm$ 이상인 확률은 22% 입니다. 그러나 아빠키가 평균이상이라는 조건이 들어가면 확률은 35%로 높아집니다.
   * 평균도 비숫한 패턴입니다. 전체 아들 평균키는 $174.5 cm$ 이나, 아빠키가 평균이상 이라는 조건이 붙으면 $177.4cm$ 로 됩니다.
   * 설명변수 (아빠키) 를 이용하여 조건을 설정하고, 계산한 확률을 **조건부 확률, 조건부 평균** 이라고 합니다.

   * 설명변수는 관심변수와 독립이 아닌 변수를 선택하여야 합니다.
       * 예를 들어 30년전 아빠의 성적은 아들 키하고는 아무런 관련이 없습니다.
       * 로또인 경우도 마찬가지 입니다. 이번주 33번이 나왔다고 다음주에 33이 나온다는 확률이 바뀌지는 않습니다.
       * 이런 관심변수와 설명변수의 관계는 **독립** 입니다.

## 심슨의 역설(Simpson's Paradox)

   * 부분에 대한 평균이 크다고 해서 전체에 대한 평균이 더 크다고 할 수는 없다
   * $\dfrac{a_1}{A_1} \gt \dfrac{b_1}{B_1}$ 이고 $ \dfrac{a_2}{A_2} \gt \dfrac{b_2}{B_2}$ 라고 $\dfrac{(a_1 + a_2)}{(A_1 + A_2)} \gt \dfrac{(b_1 + b_2)}{(B_1 + B_2)}$ 인 것은 아니다.

   * 이유는 "전체 통계를 낼 때 심슨의 역설이 발생한다면, 중요한 변수가 무시되었기 때문이거나, 각 부분의 샘플 크기나 비율이 현저히 다른데도 가중치를 주지 않았기 때문일 수 있다."
   * 통계를 조작하는 것은 아니지만. 고려하는 변수를 무시하는 경우에 발생하며, 통계를 전혀 다르게 해석 할 수 도 있다. 또한 의도를 가지고 무시 하는 경우에는 매우 심각한 결과를 초래 할 수 있는 "통계의 함정" 이라고 할 수 있다.
   
   * 간단한 예로서, 의사 A, B 가 각각 중환자 수술과 가벼운 수술만 맏는다고 한다면, 당연히 의사 A 의 수술성공율이 떨어져서 전체 평균을 떨어뜨릴 것이고, 의사 B는 당연히 수술성공률이 높아서 전체 평균을 높일 것이다. 따라서 수술의 난이도를 반영하여 정당하게 평균을 계산하지 않으면 결과는 왜곡되거나 의도적으로 나쁘게 이용될 것이다.
