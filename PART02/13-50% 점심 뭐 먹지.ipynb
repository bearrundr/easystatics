### 비교평가 그리고 선택

#### AB 테스팅

   * 보통 2가지 버젼을 만들어 놓고 임의적으로 소비자에게 노출시켜 반응을 살피는 것을 AB 테스팅이라고 합니다.
   * 예를 들어 웹쇼핑몰 에서 빨강, 파랑 구매 보턴을 만들어서 놓고 빨강, 파랑 중에서 어떤 보턴을 더 많이 누르는지 등을 알아봅니다.
   * 오바마 사례가 AB 테스팅을 잘 활용하였다고 합니다.
        * 기부보턴에 대한 적절한 색상
        * 기부 절차 단계의 흐름을 어떻게 할 지 등에 대한 적용
        * A/B 테스트 결과로 가장 나은 기부금을 이끌어 낼 수 있는 방안을 정량적인 수치로 확인 하였고, 실제 기부금 증가가 이루어 졌다고 합니다.

#### 응답 반응에 대한 평가

   * 광고 배너 안을 A, B 두개 안을 놓고, 임의적으로 사람들에게 노출 시키고 어느쪽 배너가 더 사람들에게 관심을 끄는지 반응을 분석하고자 할때, 단순 카운팅 보다는 두 안에 대한 효과 차이를 입증 해야만 합니다.
   * 분산분석 (ANOVA **AN**alysis **O**f **VA**riance) 이 유용한 도구 중에 하나가 됩니다.
   * 이번 장에서는 ANOVA 에 대한 개념을 설명 합니다.

## 분산과 분산분석

#### 분산분석 개념

   * 분산분석에 대한 개념을 먼저 다음의 예를 가지고 설명하고자 합니다.
   * 관심변수를 설명변수를 조건으로 하면 어느 정도 이유에 대해서 설명은 가능하게 됩니다.
   * 그러나 통계에서는 모든것이 다 그렇게 되지는 않습니다. 대체적인 경향은 설명이 가능하지만 하나하나 상세히 들어가면 일부는 다른 결과를 가질 수 도 있습니다.
   * 왜 그러한지에 대해서 다음의 예를 가지고 설명하고자 합니다.
   
   
   
   
   * 6명의 학생에 대한 성적입니다.
   
   * 먼저 중간 시험에 받은 점수를 수직선 상에 표현해 봅니다.

   * 기말시험은 조금 어려워서 다음과 같은 점수를 받았다고 합니다.

   * 따라서 분산이 0 이라면 차이가 없다는 뜻이고, 분산이 커지면 커질수록 차이가 크다는 의미 입니다.
   
   * 이번에는 차이가 있으므로 그 차이가 어디서 오는 것인지 알아야 합니다.
   * 그래서 **공부방법이라는 설명변수** 를 생각해 봅니다.

   * 파란색 맨 왼쪽은 회색 맨 오른쪽 보다 성적이 낮습니다.
   * 모든 파란색이 회색보다 성적이 높다고는 할 수 없습니다.
   * 아마도 맨 왼쪽의 파란색은 그날 당일 배탈이 낫거나, 아니면 기타 다른 알수 없는 영향으로 성적이 낮아졌다고 밖에 설명 할 수 없을 것입니다.

   * 정리 하면, 
   * 관심변수 차이가 0 이라면 분산도 0 입니다. 그러나 관심변수가 차이가 발생하면 따라서 분산도 차이가 발생합니다.
   * 따라서 관심변수의 차이가 있다는 정보를 분산이 가지고 있는 것이라고 생각 할 수 있습니다.
   
   
   * 그럼 분산의 차이를 설명변수를 조건으로 하면 설명이 가능하긴 한데, 이것이 완전하게 설명을 하지는 못합니다.
   * **그래서 분산은 설명이 가능한 분산 + 설명을 할 수 없는 분산 으로 나눌수 있습니다.**
   * 분산분석(ANOVA) 는 분산을 어떻게 나눌수 있는지를 간단한 수식으로 정리 했습니다.
   
   
   * 그리고 설명을 할 수 없는 부분은 그날의 컨디션, 환경등등 랜덤적 성격을 같습니다.
   
   * 따라서 설명변수가 관측치들의 차이를 얼마나 자세하게 설명하는냐 에 따라서 설명 변수의 유용성이 있겠습니다.

#### 다시 살충제 예제로 돌아 와서

##### 전체 분산 계산

   * 분산분석에서는 분산을 어떤 방법으로 나누는지에 대해서 살충제를 예로 들어 설명 합니다.
   
   * 살충제로 죽인 벌레의 전체 평균을 구해 봤습니다.
   
  <img src="../image/ed086.png" width="400" height="500" align="left">  Page 129

   * 살충제 실험에서 전체 벌레를 죽인 평균은 9.5 입니다.
   * 분산은 51.9 가 나옵니다. 각숫자에서 평균을 빼고 제곱한 다음 모두 더해 (72 - 1) 로 나누면 나옵니다.
   * 분산분석은 (72 - 1) 로 나누기 전의 숫자를 사용합니다.
   
   * 편의상 죽은 벌레수 라는 변수를 $y$ 로 하면 다음과 같은 식이 됩니다.
   
   $$\sum_{i=1}^{72}(y_i - {\bar y})^2 = \sum_{i=1}^{72}(y_i - 9.5)^2 = 3,684$$
   

##### 그룹에 따른 분산 차이 계산

   * 이번에는 각 그룹의 평균값으로 대체 해서 분산값을 계산하려고 합니다.
   * 이 변수를 편의상 $yg$ 라고 합시다.
   
   * 아래 그림을 참조하시기 바랍니다.
   
   <img src="../image/ed087.png" width="400" height="500" align="left">  Page 131 

   * 실제로 계산하면 2,669 가 나옵니다.
   
$$ \sum_{i=1}^{72}(yg_i - {\bar y})^2 \\ = 12 \times (14.5 - 9.5)^2 + 12 \times (15.3 - 9.5)^2 + 12 \ times (2.1 - 9.5)^2 + 12 \times (4.9 - 9.5)^2 + 12 \times (3.5 - 9.5)^2 + 12 \times (16.7 - 9.5)^2 = 2,669$$

##### 그룹과 개인 차이에 대한 계산

   * 실제값과 그룹평균 차이를 제곱하고 합합니다.
   * 즉 A 살충제에서서는 A 살충제 평균을 빼고, B 살충제에서는 B 살충제 평균을 빼고,... 이런식으로 합니다.
   * 그래서 계산을 하면

$$\sum_{i=1}^{72}(y_i - yg_i)^2 = 1,015$$

##### 정리

   * 이제 지금까지 계산한것을 정리하면
   * $3,684 = 2,669 + 1,015$ 가 됩니다.
   
   
   
   
   * 이것을 수식으로 표현 하면 다음과 같습니다.
   $$\sum_{i=1}^{72}(y_i - {\bar y})^2 = \sum_{i=1}^{72}(yg_i - {\bar y})^2 + \sum_{i=1}^{72}(y_i - yg_i)^2$$

   * 첫번째 항은 관심변수의 분산입니다.  즉 정보량 입니다.
   * 정보량은 두가지 부분으로 나누어 집니다.
   
   * 두번째 항이 그룹에 따른 차이 입니다.  전체 평균에 비해서 각 그룹 평균이 얼마만큼 떨어져 있는지를 나타냅니다.
   * 이값은 관심변수의 정보에서 그룹에 따른 차이를 설명하는 부분입니다.
   
   
   * 세번째 항은 실제값에서 그룹의 평균을 뺀값 입니다.
   * 그룹의 차이로 설명할 수 없는 개인 차이입니다.

   * 관심 변수 분산이 어떻게 분해 되는지 를 보았는데 이것을 분산 분석이라고 합니다.
   * 이제 그룹에 따른 차이를 어떻게 판단하는지에 대해서 논의 할 것입니다.
   * 이것은 PART 3 에서 논의 합니다.
