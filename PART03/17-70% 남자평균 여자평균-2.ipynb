#### t-통계량 히스토리

   * t-statistic, t-값, t-통계량 이란 말은 은 Student's t-tests 에서 나온것이며 Student 는 윌리엄 고셋의 필명입니다.
   * 고셋은 아일랜드 기네스 양조장에서 일하고 있었는데, 1908년에 맥주의 품질관리를 위해서 고안했습니다.
   * 기네스는 기네스북으로 유명하며, 기네스 맥주는 흑맥주로 유명합니다.
   * 기네스 회사는 맥주의 품질관리를 어떻게 하는지가 회사기밀이라고 하여 논문발표를 회사 이름으로 사용하지 못하게 하였는데, 고셋은 'Student' 라는 가명으로 자신이 고안한 분포를 분석한 논문을 발표 했습니다.
   * [t-통계량에 대한 위키](https://en.wikipedia.org/wiki/T-statistic)
   
   
   * [t-분포에 대한 위키 한글](https://ko.wikipedia.org/wiki/%EC%8A%A4%ED%8A%9C%EB%8D%98%ED%8A%B8_t_%EB%B6%84%ED%8F%AC)
   
   
   
   * 프리드리히 로베르트 헬메르트(독일어: Friedrich Robert Helmert)가 1875년에 도입하였다. 이듬해 야코프 뤼로트(독일어: Jacob Lüroth)도 같은 분포를 재발견하였다. 그러나 헬메르트와 뤼로트의 논문은 영문 학계에 널리 알려지지 않았다. 1908년에 윌리엄 고셋이 "스튜던트"(영어: Student)라는 필명으로 1908년 재발견하였다. 고셋은 기네스 양조 공장에서 일했고, 맥주에 사용되는 보리의 질을 시험하기 위해 이 분포를 도입하였고, 경쟁사들에게 기네스의 획기적인 통계 기법을 숨기기 위해 필명을 사용하였다고 한다. 이후 저명한 통계학자인 로널드 피셔는 이 분포를 "스튜던트 분포"라고 불렀고, t라는 기호를 사용하였다. 피셔 이후 이 분포는 고셋의 필명을 따 "스튜던트 t 분포"로 알려지게 되었다.

#### t-statistic 사용 용도

   * t-statistic (t-통계량)은 표본평균이 0 인지 아닌지 판단할때 사용합니다. 
   * 표본평균이 0 이 아닐때도 사용합니다.
   
   
   * 통계 가설 검증 형식으로 보면 귀무가설 $H_0: \beta = \beta_0$ 형식이며 $\beta_0$ 가 0 이거나 0 이 아닐때 사용합니다.
   * t-통계량은 pivotal-quantity(중심추축량) 이라고 합니다. pivotal-quantity 는 식에는 미지의모수를 포함하고 있지만 그것의 분포에서 포함하지 않는 통계량을 말합니다. 실제 다음에 나올 식에서 보면 미지의 모수 평균이 들어가 있으나 실제 t-분포에서는 그 값이 없습니다. 이런 것을 pivotal-quantity 라고 합니다. 이 성질을 이용하여 모수의 평균에 대한 신뢰구간을 구하는데 사용합니다.
   
   
   * 관측치가 많은 경우에는 정규분포를 사용하면 되나, 실제에 있어서는 많은 관측치를 구하기가 쉽지 않습니다. t-값은 적은 관측치를 가지고 검정을 하는데 정규분포보다 정확하다고 하여 많이 사용합니다.

   * 뺀 결과 다음과 같습니다.
   
 <img src="../image/ed109.png" width="500" height="200" align="left">  Page 174

  * 결과를 보면 가장 낮은 도수는 기준값에서 0.05% 낮고, 가장 높은 도수는 기준값에서 0.07% 입니다.

##### 2단계 이렇게 구한 값들을 이용하여 표본평균과 표본편차 및 표본평균 표본편차(표본에러 라고도 합니다) 를 구합니다.

   * 표본 $x$ 의 평균 ${\bar x}$ 는 0.01% 가 됩니다.
   
$$
{\bar x} = \dfrac{(-0.05\%) + (-0.01\%) + (0.01\%) + (0.03\%) + (0.07\%)}{n} = \dfrac{0.05\%}{5} = 0.01\%
$$

   * 표본분산 $s^2_x$ 는 0.02 가 됩니다.
   
$$
s^2_x = \dfrac{(-0.05 - 0.01)^2 + (-0.01 - 0.01)^2 + (0.01 - 0.01)^2 + (0.03 - 0.01)^2 + (0.07 - 0.01)^2 }{5 - 1} = \dfrac{0.008}{4} = 0.002
$$

   * 표본의 평균 알코올 도수는 4.21% 입니다. 이제 남은 표본평균의 표준편차(표준에러 라고도 함) 을 구하여여 하는데, 표본의 크기가 5 이므로  단순히 표본분산을 표본크기로 나누어 sqrt 하면 됩니다.
   
$$
s_{\bar x} = \sqrt{\dfrac{0.002}{5}} = 0.02\%
$$

##### 3단계 t-통계량(statistic) 또는 t-값 구하기

$$
t-값 = \dfrac{표본평균 - 기준값}{표본평균의 표분편차(또는 표준에러)}
$$

   * 계산을 해보면 t-value = 0.5 가 됩니다.
$$
t-value = \dfrac{0.01}{0.02} = 0.5
$$

#### 참고로 위키에서 식 표현은 

[t-statistics](https://en.wikipedia.org/wiki/T-statistic)

$$
t_{\hat \beta} = \dfrac{{\hat \beta - \beta_0}}{s.e.({\hat \beta})}
$$

   * ${\hat \beta}$ : 통계모델에서 모수 $\beta$ 의 추정치 (estimator) 입니다.
       * 위 식에서 표본평균입니다. 표본평균은 모집단에서 모수가 될 수는 없습니다. 그러나 표본을 추출하여 구한 값이므로 이 값이 모집단의 모수에 근접할 것이라고 추정합니다.
   * $\beta_0$ : 상수이고 기준값입니다. 이것은 미모수 $\beta$ 에관계 있을수도 없을수도 있습니다. 기본은 $\beta_0$ = 0 입니다. 그러나 필요에 따라서 $\beta_0$ 는 0 이 안될수도 있습니다. 이럴경우 귀무가설 $H_0: \beta = \beta_0$ 에서 $\beta_0$ 은 0 이 아닙니다.
   
   * $s.e({\hat \beta})$ : standard error 이며 지금까지 설명한 표본평균의 표본편차입니다.

#### t-값의 이용

   * t-값은 표본으로 가지고 있는 데이터가 기준값으로 부터 얼마나 떨어져 있는지를 나타냅니다.
   * 따라서 이 t-값이 t-분포에서 상대적인 위치를 확인 하여 적절한 값인지 아닌지를 테스트(검정) 하게 됩니다.

## t-분포

   * 60% 지구는 우주의 티끌 에서와 같이 평균차이가 0이라는 가정에 표본을 추출하면 t-분포가 만들어 집니다.
   * 사실은 t-value 를 구하는 것에서 부터 표본 추출하는 과정을 진행하면 t-분포가 만들어 집니다. 이과정은 책의 범위를 벋어나므로 필요시 통계책을 보기로 하구요
   * 여기서는 t-분포를 활용하는 방법에 대해서 논의 하려고 합니다.

   * t-분포식을 살펴보면 다음과 같습니다.
   
$$
f(x) = \dfrac{\Gamma(\dfrac{\nu + 1}{2})}{\sqrt{\nu \pi}\Gamma(\dfrac{\nu}{2})}\left(1 + \dfrac{x^2}{\nu}\right)^{-\dfrac{\nu+1}{2}}
$$

   * 여기서 관심은 모수(Parameter) 를 찾는것입니다. t-분포에서 모수는 $\nu$ 입니다. 자유도 (degree of freedom) 이라고 하는데, 관측갯수 - 1 을 말합니다.
$$ \nu = n - 1 $$

   * 관측치 에서 1을 뺴는 이유는 이미 평균에서 값을 결정했기 때문에 관측치에 1 을 뺀다고 볼 수 있습니다.
   * 자유도란 쉽게 말하면 미지의 변수 갯수입니다. 여기서는 관측치 갯수가 되겠습니다.
   * 그런데 제한조건을 주면 그 순간 제한조건 만큼 자유도는 줄어들게 됩니다.
   * 예를 들어 3개의 변수 $x, y, z$ 가 있다고 해요. 그럼 3개의 변수는 아무런 값을 선택하여 넣을수 있습니다. 그런의미에서 자유도는 3입니다.
   * 이제 $x + y + z = 10$ 이라는 조건을 두면, x, y 가 어떤값을 갖게 되면 z 는 그 순간 자기가 아무런 값을 가질수 없습니다.왜냐하면 저 식이 구속조건이 되서 반드시 합이 10 되어야 하기 때문입니다. 이런 의미에서 자유도는 2 입니다.
   
   
   * 마찬가지로 관측수를 다 더해서 평균이라는 것을 알고 있기 때문에 자유도는 관측치 수에서 1을 빼야만 합니다.

   * 맥주 예제에서 관측치가 5 이므로 자유도는 4가 됩니다. 이것을 그래프로 그리면 다음과 같습니다.
   
 <img src="../image/ed110.png" width="500" height="200" align="left">  Page 177

   * t-분포에서 평균은 0 입니다. 좌우 대칭이 종모양으로 정규분포 하고 비숫합니다. 정규분포에 비해서 좌우가 약간 올라가 있습니다.
   * 관측치가 많을 경우에는 정규분포를 사용하나, 적은 경우에는 t-분포가 적은경우에 적용 할 수 있다고 합니다. 사실 여기서 관심영역은 좌우 끝트머리 이기 때문입니다.

   * 맥주공장에서 하루개 표본으로 수십개씩 뽑아 사용할 수는 없기 때문에 고안 한 것이라고 할 수 있겠습니다.

```r
# 다음은 정규분포하고 비교하기 위해서 표준정규분포하고, nu = 2,4,10 인 경우 각각을 그려 본것입니다.
# 보면 정규분포가 좌우 끝트머리에서 제일 아래 부분에 있죠. 그리고 2 일 경우가 가장 높게 나타납니다.

# 먼저 테스트 데이터를 생성합니다.
x=seq(from=-5, to=5, by=0.01)

# 정규분포를 그립니다.
plot(x, dnorm(x), type='l', col="red", xlab="x", ylab="", main="Normal, T-Distribution")

# 추가 해서 t=2 를 합니다.
lines(x, dt(x,2), lty=2, col=3, lwd=3)

# t=4
lines(x, dt(x,4), lty=3, col=4, lwd=3)

# t=10
lines(x, dt(x,10), lty=4, col=5, lwd=3)

# 마지막으로 범례를 표시합니다.
legend(3,0.4, c("N(0,1)", "t(2)",  "t(4)", "t(10)"), col=c(2,3,4,5), lty=c(1,2,3,4), lwd=c(1,3,3,3))
```

2. 계산된 t-값을 자유도가 n-1 인 t-분포에 넣어 p-값을 계산하고 유의수준과 비교하기 입니다.

   * t-검정을 역으로 활용해서 신뢰구간을 구하는 방법에 대해서 논의 합니다.
