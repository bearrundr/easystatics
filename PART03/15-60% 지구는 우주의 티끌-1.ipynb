## 표본과 모집단

#### 샘플과 모집단

   * 모래 백사장을 생각해 봅시다.
   * 거기서 한줌 퍼 올리면 작은 모래 알갱이가 만 개쯤 들어 옵니다.
   * 손을 털고 다시 모래를 퍼 올리면 또 다른 만개의 모래알이 손안에 있습니다.
   * 백사장에 있는 수억개의 모래알중 만여개가 손안 한줌에 있습니다.
   * 이런것을 **표본(Sample)** 이라고 합니다.
   * 당연히 모래 백사장은 **모집단(Population)** 이라고 합니다.
   * 모래에서 퍼 올린 한줌은 하나의 **표본** 이 됩니다.

#### 표본이 필요한 이유

   * 지금까지 PART 1, PART 2 에서는 데이터의 차이를 구하는 방법에 대해서 논의 하였습니다.
   * 평균은 구하여 관측치와 비교를 하고, 교차표를 이용해서 각 데이터의 비중(올림픽 메달종류별 비교참조)을 확인하고, 살충제 효과를 구하기 위해 살충제별 평균을 구하고 차이를 확인하였습니다.
   * 그러나 이런 차이점이 당현한 것인지 아니면 필연인 것인지 어떤 의미가 있는 것인지 알 수 가 없습니다.
   * 따라서 이런 차이가 의미가 있으려면 비교대상이 있어야 하고 이문제를 표본을 통해서 확인합니다.
   * 표본을 통해서 모집단의 모습을 예측하고 구해진 데이터의 차이가 그 모집단의 모습에서 어디에 속하는지를 파악하는 것입니다.

## 통계량과 분포

  * 이번에 데이터의 분포에 관해서 이야기 하고자 합니다.
  * 역시 백사장의 모래 이야기 입니다.
  * 어떤 섬에서 가지고 온 모래 한 줌이 있습니다.
  * 그리고 그 모래 한줌이 곱다고 주장합니다.
  * 그 모래가 곱다고 주장에 반론을 할려면 가까운 해변에서 모래 한줌을 가지고 와 비교를 하면 됩니다.
  * 한줌의 모레를 가지고 와서 모래알의 평균 굵기를 구하고 비교를 하여 곱다고 한 모래의 평균 굵기와 비교를 하면 됩니다.
  
  
  * 통계에서는 비교를 하기위해 다음과 같은 방법을 사용합니다.
  * 해변에서 가지고 온 모래 표본들의 평균 모래 굵기의 패턴을 찾으려고 합니다.
  
  
  
  * 그러나 해변에서 가져온 모래한줌의 평균 굵기가 항상 일정한지 먼저 살펴 보고자 합니다.
  * 지금 곱다고 한 한줌에 있는 모래 굵기는 3 입니다. 따라서 백사장에서 가져온 모래 굵기가 평균적으로 항상 3보다 크다면 모래가 곱다고 한것은 중명이 됩니다.
  * 모래 한줌은 약 10,000개의 모래가 있습니다.

## 자연스러운 확률

#### 아빠키와 아들키의 상관관계 (연속형변수)

   * 앞서 아빠키와 아들키의 상관계수가 0.5 가 나왔습니다. 
   * 이값이 의미가 있는 것인지 알아보고자 합니다.
   
   * 역시 표본을 이용한 방법입니다.
   * 먼저 전혀 관련이 없는 두 변수의 상관계수를 구해보도록 하겠습니다.
   * 이들의 상관계수가 0 이 나올까요? 평균은 0 이 나올것입니다. 그러나 항상 0 이나오지는 않습니다.
   
   * 상관계수를 구하기 위해서 두 변수를 랜덤으로 만들도록 하겠습니다.
   * 첫번째 변수는 평균이 171.9 에 표준편차가 7이 되도록 랜덤으로 1078개의 관측치를 만들었습니다. (아빠키 에 해당 하는 변수)
   * 두번째 변수는 평균이 174.5 에 표준편차가 7이 되도록 랜덤으로 1078개의 관측치를 만들었습니다. (아들키 에 해당 하는 변수)
   
   * 두변수 모드 관측치 수가 1078개 이므로 하나씩 짝을 지워봅니다.
   
  <img src="../image/ed095.png" width="500" height="300" align="left">  page 147

   

   * 이런것을 임의(Random) 로 만든 표본이라고 해서 임의 표본이라고 합니다.
   * 전혀 관련이 없는 두 변수를 아빠, 아들같이 하나로 묶어서 상관계수를 계산합니다.
   * 그 결과 0에 가까운 0.01 이 나왔습니다.
   * 한번의 랜덤샘플로는 의미가 없으므로 상관계수에 대한 분포를 만들어 봅니다.
   
   
   * 10,000 번 반복해서 10,000개의 랜덤 샘플을 만들고 그리고 나서 10,000 개의 상관계수를 히스토그램으로 그렸습니다.
   
  <img src="../image/ed096.png" width="400" height="200" align="left">  page 148


   * 대부분의 상관계수는 0 이지만 10,000 개의 상관계수 중에서  $\pm 3$ 을 넘는것이 22 개 가 있습니다. 0.22%의 확률로 일어나고 있습니다.
   * 전혀 상관없는 두변수로 계산한 상관계수가  $\pm 3$ 인 것은 통계에선 0.22% 의 확률로 일어나는 사건이라고 봅니다. 즉 우연의 정도를 측정해서 확률로 표현합니다.
   
   
   * 아빠키와 아들키의 상관계수는 0.5 입니다.
   * 상계계수 분포에서보면 $\pm 0.5$ 이상에서는 한번도 나오지 않았습니다.
   * 아마 횟수를 1억번 정도 올리면 한번 정도 나올 수 있을 것같습니다.
   
   
   * 즉 발생할 확률이 0.0000001% 정도 되는 매우 특이한 일입니다.
   * 따라서 아빠키와 아들키는 상당히 의미 있는 양의 상관관계가 있다고 할 수 있습니다.
