##### 페이지 189 교차표를 살펴보면

   * 같은 방법으로 나머지 11칸도 숫자들이 계산 되어 있습니다.
   * 만일 4번의 대회와 메달수가 전혀 관련이 없다면 실제값과 예상값이 차이가 없을것이며 12개의 차이는 대부분이 0이 될 것입니다.
   * 만일 차이가 많다면 두 변수 사이에는 뭔가가 있을 것 같습니다.
   * 즉 범주형 변수의 수준간에 차이가 있을 것 같습니다.

##### 범주형 변수간의 유의성 검정

   * 교차표에 있는 12개 숫자를 활용해서 차이를 의미하는 것을 만들어 내고자 합니다.
   * 일단 표에서 차이가 없는 경우 0 이 됩니다.
   * 음수 양수를 다 고려 해야 하므로 
   * 일단 분산을 구하는 것 처럼 제곱을 합니다.

   <img src="../image/ed117.png" width="500" height="200" align="left">  Page 190

   * 분산을 구하는 것처럼 바로 합하는 것이 아니라 이값을 예상값으로 나눕니다.
   
   * 아테네의 경우 실제 메달수는 9 예상값은 12 이므로 $(9-12)^2 = 9$ 입니다. 예상 메달수는 12 입니다.
   * 따라서 9 를 12 로 나누면 0.75 가 됩니다.
   

$$
\dfrac{(실제값-예상값)^2}{예상값} \qquad = \dfrac{(9-12)^2}{12} = \dfrac{9}{12} = 0.75
$$

   <img src="../image/ed118.png" width="500" height="200" align="left">  Page 191

   * 이제 12개의 값을 모두 더합니다.
   
   0.75 + 0 + 0.36 + 0.13 + 1 + 0 + 0 + 1.5 + 0 + 0.1 + 0.44 + 1.5 = 5.78
   

   * 이렇게 계산된 차이 정도의 합계를 **카이제곱통계량(Chi-squared statistic)** 또는 **카이제곱값(Chi-squared value)** 라고 합니다.
   * 어떤 변수를 제곱한값 정도 의미 입니다.
   * 제곱을 해서 더한값은 반드시 평균이 0 이고 표준편차가 1인 표준 정규분포를 따라야만 합니다.
   * 그이유가 먼저 예상값으로 나눈 이유이기도 합니다.

이것을 수식으로 나타내면 다음과 같습니다.
$$
\chi^2 = \sum\dfrac{(O_i - E_i)^2}{E_i}
$$

$O_i$ : 관측치

$E_i$ : 예상치

##### 유의성 검정 과정

만일 $\chi^2$ 이 0 이라면 ??

   * 반대로 값이 커지면 커질수록 실제 교차표가 독립인 상황과는 아주 다르다고 할 수 있습니다.
   * 따라서 독립인지 아닌지를 판단할 기준이 필요합니다.
   * 카이제곱 분포를 구해서 , 분포에서 $\chi^2$ 의 p-값을 구하고 역시 유의성 검증을 합니다.

##### 카이제곱분포(Chi-squared distribution)

카이제곱분포 식은 다음과 같습니다.
$$
f(x) = \dfrac{1}{2^{\dfrac{k}{2}}\Gamma(\dfrac{k}{2})}x^{\dfrac{k-1}{2}}e^{\dfrac{k}{2}}
$$

카이제곱 분포는 감마분포의 특수한 형태입니다. 즉 감마분포 $\alpha$, $\beta$ 에서 $\alpha = \dfrac{k}{2}$, $\beta = \dfrac{1}{2}$ 인 특수한 경우 입니다.

카이제곱 분포의 모수는 k 이고 자유도의미 입니다. 자유도는 (대회별 - 1) x ( 메달별 - 1) 입니다. 즉 교차표에서 가로 세로 각각 1을빼고 곱하면 나옵니다.
즉 $(4-1)\times(3-1) = 6$ 입니다. 다음 그림에서 파란색이 자유도 입니다.

   <img src="../image/ed119.png" width="500" height="200" align="left">  Page 193

```r
# 다음은 Chi-Squared 를 그려본것입니다.Chi(6) Chi(10), Chi(20), Chi(30)

# 먼저 테스트 데이터를 생성합니다.
x=seq(0,20, length=200)

# 자우도 = 6
plot(x, dgamma(x,6/2,1/2), type='l', col="red", xlab="x", ylab="", main="Chi-Squared-Distribution")

# 자유도=8 
lines(x, dgamma(x,8/2,1/2), lty=2, col=3, lwd=3)

# 자유도 = 10
lines(x, dgamma(x,10/2,1/2), lty=3, col=4, lwd=3)

# 자유도 = 12
lines(x, dgamma(x,12/2,1/2), lty=4, col=5, lwd=3)

# 마지막으로 범례를 표시합니다.
legend(15,0.1, c("Chi(6)", "Chi(8)",  "Chi(10)", "Chi(12)"), col=c(2,3,4,5), lty=c(1,2,3,4), lwd=c(1,3,3,3))
```

#### 독립성 검정

   * 교차표를 활용해서 두 범주형 변수가 독립인지 아닌자 판단하는 방법을 **독립성 검정(Test of Independence)** 라고 합니다.
   
   
   * 선거출구조사에서 지역, 성별, 연령대, 그리고 누구에게 투표했는지 기록하는데 이런것들이 범주형 변수가 됩니다.
   * 각 후보별 지역별, 성별, 연령별 득표수를 계산 할 수 있으며 모두 교차표로 표현이 가능합니다.
   * 카이제곱을 활용해서 두 변수의 독립성검정을 하면, 각 후보를 지원하는 사람들의 성향이 우연히 조금 다르게 나온것인지 아니면 우연이라고 하기에는 정말 큰 차이를 보이는지를 판단 할 수 있습니다.
