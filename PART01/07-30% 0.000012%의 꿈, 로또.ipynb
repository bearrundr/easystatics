## 30% 0.000012%의 꿈, 로또 (7일차)

   * 로또를 이용하여 확률의 개념을 소개하고, 확률이 데이터 분석에서 활용하는 방안에 대해서 설명한다.
   * 통계는 데이터 분석에 확률을 사용합니다. 이장에서는 확률에서 사용하는 기본용어를 설명하고자 합니다.

### 일상에서 확률

   * 우리 생활에 가정 먼저 생각나는 확률은 일기예보 입니다.
   * 내일 비가 올 확률은 80% 입니다. 그럼 우산을 챙기는 것이 날것입니다.
   * 100퍼(100%)라는 것은 분명한 확신이고 0퍼(0%)는 발생하지 않을것이라는 것을 알고 있습니다.
   * 확률에서 관심 많은 로또에 관한 이야기

   * 1등은 45 개 숫자 중에서 6개의 숫자를 맞추는 로또 1등확률은

  * $\dbinom{45}{6}$ = ${}_{45}C_{6} = 8,145,060$ 개의 번호 중에서 꼭 하나는 나오므로
  * $\dfrac{1}{8,145,060} = 0.000012 \%$ 가 됩니다.

   * 2등은 당첨번호 5개와 보너스 번호를 맞혀야 하므로 모두 ${}_{6}C_{5} = 6$ 가지 입니다. 선택되면 한가지 번호로 선택이 됩니다.
   * $\dfrac{6}{8,145,060} = \dfrac{1}{1,357,510}$ 

   * 3등은 6개중 5개의 숫자가 맞고, 나머지 하나는 1 - 39 개 숫자중 아무거나, 그리고 2등 당첨 숫자 제외
   * ${}_{6}C_{5} \times (39 - 1) = 228, \dfrac{228}{8,145,060} = \dfrac{1}{35,714}$ 

   * 4등은 6 개의 숫자중 4개가 맞고,  1 - 39 개 숫자중 2개가 맞는 경우
   * ${}_{6}C_{4} \times {}_{39}C_{2} = 11,115, \dfrac{11,115}{8,145,060} = \dfrac{1}{732.799} =  0.14 \%$

   * 5등은 6개중 3개가 맞아야 하고 나머지 3개는 1 -39 개 숫자중 3개가 맞아야 함.
   * ${}_{6}C_{3} \times {}_{39}C_{3} = 18,720, \dfrac{18,720}{8,145,060} = \dfrac{1}{44.562} = 2.2 \%$

## 확률

   * 아직 일어나지 않은 미래에 일어날 불확실성 을 통계에서는 **사건(Event)** 라고 합니다.
   * 확률은 이 사건이 현실화 될 가능성을 0 에서 1 사이 숫자로 표시합니다.
   * 이 숫자를 100분율로 표시하면 $\%$ 로 표시가 됩니다.
   * 확률 $0\%$ 는 불가능을 표시하고, $100\%$ 는 필연입니다.
   * 나머지 소숫점 예를 들어 $75%$ 라고 하면 똑같은 상황이 100번 반복되었을때 75번 정도는 사건이 발생하고 나머지 25번 정도는 사건이 발생하지 않는다는 의미 입니다.


### 이론적 확률

   * 수리적 또는 논리적으로 계산한 확률로 누구나 항상 동일한 값이 나온다.
   * 실제로 발생할 사건을 모든 발생할 사건으로 동일하게 나눈값 입니다.
   * 예를 들면 동전의 앞변이 나올 확률은 $\dfrac{1}{2}$ 입니다.
   * 이론적 확률은 모두가 하지 않습니다.
       * 왜냐하면, 논리적으로 설명할 수 없는 경우도 있고, 누군가는 논리적 확률을 믿고 싶지 않기 때문입니다.


### 경험적 확률

   * 진짜 통계쪽의 주장입니다. 이런 주장을 빈도론자(Frequentist) 라고 합니다. 통계에서는 아주 중요한 개념입니다.
   * 이론적 확률보다는 실제 시행을 해보고 나온 결과를 확률로 정합니다.
   * 예를 들면, 두사람이 가위, 바위, 보를 해서, 이길 승률이 이론적으로 $50\%$ 라고 하면 진짜 가위, 바위,보를 해서 사건을 관찰한 루에 승률을 정합니다.
       * 한 사람이 일반적으로 10번 해서 10 번 이기면 승률은 $50\%$ 가 아니라 $100\%$ 입니다.
       * 동전 던지기를 하여 앞면이 10 번 시행중에 10번이 나오면 앞변이 나올 확율은 $100\%$ 입니다. 
       * 믿을수 없어서 더 시행을 하면 그럼 그때 앞면이 나온 빈도수가 확률이 됩니다.

## 데이터 분석과 확률

   * 데이터 분석에서 확률의 역활

   * 통계는 과거를 기록한 테이터 속에서 패턴을 찾고, 어떤 가능성을 확률로 표현합니다.

#### 첫번째

   * 어떤 관심 있는 확률이 궁금할때

   * 하는 일마다 성공하는 사람, 손만대면 실패하는 사람 
   * 통계에서는 성공하는 사람, 실패하는 사람의 정보를 모아 데이터를 만들고, 어떤 차이가 성공과 실패를 가르는지 **확률모형(Probability Model)** 을 만든다.
   * 모형에서 성공할 확률에 영향을 미치는 변수가 무엇인지 살펴본다.

#### 두번째

   * 어떤 차이가 있는지 궁금할때

   * 어떤 차이가 우연에 의한 것인지 아니면 필연에 의한 것인지 따져 볼 필요가 있는지 경우를 따져 볼 필요가 있다.
   * 예를 들어
      * 주변의 친구 8명을 대상으로 남자 4명중 2명, 여자 4명중 1명이 안경을 쓰고 있다.
      * 남자 여자 안경쓴 차이가 25% 라면 의미를 찿기는 어렵다.
      * 예를 들면 남자의 50% 가 안경을 쓰고, 여자의 25% 가 안경을 쓰고 있다면 이 25%의 차이에 대한 의미를 따져봐야 할 것입니다.

   * 관심있는 확률들과 그들사이의 관계를 활용해서 변수속 차이를 확률로 설명하고, 성명된 차이가 의미가 있는지 없는지를 판단하는데 통계를 사용한다.
