## factor() 함수로 범주형 변수 만들기

범주형 변수의 수준은 일반적으로 문자 형식으로 되어 있습니다. 이것을 숫자형으로 표시하면 처리 하기가 더 간편하게 됩니다. 
   * 예를 들어 성별 을 표시할때, 남자, 여자 보다는 1은 남자, 2는 여자로 코드집화하여 데이터에는 1, 2 로 표시하는 것이 편합니다.
   * 그러나 R 에서는 숫자로 표시되면 숫자 형식으로 인식하기 때문에 문제가 있을수 있습니다.
   * 이 경우 factor() 함수를 이용하여 숫자가 아닌 범주형 변수 라는 것을 R 에 인식시켜향 합니다.

```r
demo = read.csv("Practice/data/gender_age.csv")
```

```r
demo
```

```r
# 데이터의 구조를 파악합니다.
str(demo)
```

   * Gender 를 숫자로 인식하고 있습니다. 이것을 범주형 변수로 인식하게 하여야 합니다.

```r
demo$Gender
```

```r
# 범주형 변수로 인식
factor(demo$Gender)
```

```r
str(factor(demo$Gender))
```

### $ 를 활용한 변수 수정과 추가하기

```r
# Gender 를 변수형 변수로 인식
demo$Gender = factor(demo$Gender)
```

```r
str(demo)
```

```r
# 데이터 프레임에 변수 추가 하기
paste0(demo$Gender,"_",demo$Age,"대")
```

```r
# 이것을 새로운 변수 "Group" 로 추가합니다.
demo$Group = paste0(demo$Gender,"_",demo$Age,"대")
```

```r
# demo 의 구조를 보면
str(demo)
```
