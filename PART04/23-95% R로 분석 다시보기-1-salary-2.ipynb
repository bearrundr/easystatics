## 내 연봉은 어느 정도 순위에 있을까?

한국 납세자 연맹(koreatax.org) 에서는 연말정산 데이터를 기반으로 연봉 탐색 서비스를 제공합니다. 이것을 바탕으로 가상의 기업을 만들어 10,000 명이 근무하는 회사의 연봉 데이터를 만들었습니다.

```r
data = subway = read.csv("Practice/data/salary_data.csv")
```

```r
salary = data$salary
```

```r
# 평균, 분산 합계 등등
sum(salary)
```

```r
# 평균
mean(salary)
```

```r
# 분산
var(salary)
```

```r
# 표준편창
sd(salary)
```

 내 희망 연봉이 4,000 만원이면 10,000 명중 상위 몇 %나 되는지 구해 보아요


```r
head(salary > 4000, 50)
```

```r
# TRUE 는 1, FALSE 0 과 같으므로, TRUE 를 다 더하면
sum(salary > 4000)
```

```r
mean(salary > 4000)
```

   * 4,000 만원은 상위 27% 정도 되는 연봉입니다.

이제 다섯숫자를 boxplot 으로 표현해 보아요.

```r
boxplot(salary, range=100)
```

   * 아래쪽 75% 는 서로 연봉이 비숫한 수준이지만 많은 사람들이 뺵빽히 모여 있습니다.
   * 상위 25%의 연봉은 큰차이로 높다는 것이 나타납니다.
   * range = 100 옵션을 지우고 보면 다음과 같습니다.
   

```r
boxplot(salary)
```

   * 최고 높은 가로선이 1억 언저리에 있고 그 위선은 점으로 표현 되어 있습니다.
   * 이상치 (Outlier) 는 1억 보다 높은 것으로 처리 했습니다. 
   
   
   * 이번에는 1억 미만의 관측치만 처리 해 보아요


```r
salary_10000 = salary[salary < 10000]
```

```r
boxplot(salary_10000)
```

#### 연봉을 히스토그램으로 표시

   * 세로축은 사람수, 가로축은 연봉으로 표시합니다.

```r
hist(salary)
```

   * 대부분의 사람이 2,000 만원 보다 낮은 연봉을 받기 때문에 0 에 가까워 질수록 막대그래프가 높게 나타납니다.
   * 1 억까지 구간을 좀더 잘게 나누어서 보아요


```r
hist(salary_10000, breaks=seq(0,10000,500))
```

   * 특이한 점은 $0 \sim 500만원$ 구간이 $500 \sim 1,000만원$ 구간 보다 높은데, 이것은 중도 입사자, 퇴사자의 영향으로 보입니다.
